```python
# ðŸ“Œ Cell 1: Import libraries & categories
import json

CATEGORIES = {
    "Flood": ["flood", "water entering", "water logging", "water coming inland"],
    "High Wave": ["high wave", "high tide", "rough sea", "strong tide", "sea level very high"],
    "Tsunami": ["tsunami", "earthquake", "alert issued"],
    "Cyclone": ["cyclone", "storm", "wind warning"],
    "General Alert": ["alert", "siren", "situation serious", "evacuated"]
}
```

```python
# ðŸ“Œ Cell 2: Define classifier function

def classify_post(text):
    text_lower = text.lower()
    for category, keywords in CATEGORIES.items():
        if any(keyword in text_lower for keyword in keywords):
            return category
    return "Uncategorized"
```

```python
# ðŸ“Œ Cell 3: Load your social_media.json

input_path = r"C:\Users\hp\Downloads\Project\data\live_social_media.json"

with open(input_path, "r", encoding="utf-8") as f:
    data = json.load(f)

print(f"âœ… Loaded {len(data)} posts from {input_path}")
data[:2]  # preview first 2 posts
```

```python
# ðŸ“Œ Cell 4: Classify each post

for post in data:
    post["type"] = classify_post(post["text"])

print("âœ… Classification Done")
data[:3]  # preview first 3 classified posts
```
